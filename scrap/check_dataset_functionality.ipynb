### Make sure new dataset is working correctly

```python
import sys
sys.path.append("..")
from utils.data_loaders import DataLoader
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
dl = DataLoader(dataset='MNIST_interps', tr_batch_size=1, te_batch_size=1, 
                augment=False, model='lenet', path='../data', thresh=.75)
train_loader = dl.train_loader
test_loader = dl.test_loader
```

```python
sample,label,target_interp = next(iter(train_loader))

fig = plt.figure(figsize=(6,3), constrained_layout=True)
spec = gridspec.GridSpec(ncols=2, nrows=1, figure=fig)

ax1 = fig.add_subplot(spec[0,0])
ax1.title.set_text('sample')
plt.imshow(sample.squeeze().numpy(), cmap='cool')
plt.colorbar()

ax2 = fig.add_subplot(spec[0,1])
ax2.title.set_text('target_interp')
plt.imshow(target_interp.squeeze().numpy(), cmap='cool')
plt.colorbar()

plt.show()
```
